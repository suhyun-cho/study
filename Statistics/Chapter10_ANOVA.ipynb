분산분석 (analysis of variance, ANOVA)은 두 개 이상 다수의 집단을 비교하고자 할 때 집단 내의 분산, 총평균과 각 집단의 평균의 차이에 의해 생긴 집단 간 분산의 비교를 통해 만들어진 F분포를 이용하여 가설검정을 하는 방법이다.  <br>
즉, 순수한 오차에 의한 영향보다 큰 영향을 주는 요인이 어떤 것인가를 규명하는 분석방법. 

* 평균 차이를 비교하는데, 분산의 개념을 활용하는 이유
    * 집단의 평균들이 멀리 떨어져 분산이 크면 클수록 집단간의 평균들이 서로 다르기 때문이다. 

1) 일원분산분석(one-way ANOVA) 
    * 종속변인은 1개이며, 독립변인의 집단도 1개인 경우
2) 이원분산분석 (two-way ANOVA) 
    * 독립변인의 수가 두 개 이상일 때 집단 간 차이가 유의한지를 검증하는 데 사용

[Q] : <br>
t-test를 여러번 하지 않고 ANOVA테스트를 해야하는 이유 ? <br>

[A] : <br>
여러번 중복해서 비교할수록 신뢰도가 하락하기 때문에. <br>
        예를들어, t-test 알파 0.05로 비교실험을 했을때 신뢰도는 (1−α) 이다. 이러한 실험을 두번 반복했다고 할때, 통계학에서는 두개의 사건이 동시에 일어나는 경우 곱으로 처리힌다. <br>
        따라서 이때의 신뢰도는  (1−α)2 가 됩니다. 한번 비교했을때보다 신뢰도가 감소함.

## 10.1 일원분류 분산분석

<용어>
* 인자(factor) : 어떤 실험에서 관측값에 영향을 주는 조건의 종류
* 인자수준 (factor level) : 실험을 하기 위한 인자의 여러가지 조건 <br>
ex) 인장강도가 반응온도에 따라 어떻게 변하는가를 비교하고자 할 때에 온도는 '인자', 60도, 70도, 80도 의 온도에서 실험을 했다면 이 값는 '인자수준' 이다.

* 처리 제곱합(treatment sum of squares) : 요인의 변화에 따른 반응의 차에 의한 제곱합 
* 잔차 제곱합(residual sum of squares) : 아직 원인이 규명되지 못한 요인에 의한 제곱합

[일원분류의 특징] <br>
(1) 각 수준에서의 반복수는 같지 않아도 좋으며, 많이 사용되는 수준수는 3~5, 반복수는 3~10이다. <br>
(2) 실험의 측정은 랜덤하게 선택된 순서에 의하여 독립적으로 시행되어야 한다. 

[일원분류의 모집단 모형] <br>
### $Y_{ij} = \mu + \alpha_i + \epsilon_{ij}$ (관측치 = 전체평균 + 처리효과 + 오차)
#### = $Y_{ij} = \mu_i + \epsilon_{ij}$ $(\alpha_i = \mu_i - \mu)$

($i = 1, ..., k,  j=1, .., n_i$) <br>
$\mu $ : 총평균 <br>
$\alpha_i$ : i번째 처리효과로서 $\sum_i^k \alpha_i = 0$ <br>
$\epsilon_{ij}$ : 오차항으로서 서로 독립인 $N(0,\sigma^2)$ 확률변수 <br>

(오차항은 서로 독립인 확률변수)
* 정규성 : 오차항은 정규분포를 따른다.
* 독립성 : 모든 오차항은 서로 독립이다.
* 불편성 : 오차의 기대값은 0이고 편의는 없다.
* 등분산성 : 모든 i,j에 대하여 오차항의 분산은 $\sigma^2$ 이다.

<img src="ANOVA.PNG" width="700"> <br>

I개의 처리효과에 차이가 있는지를 검정하는 것은 I개의 모평균 사이에 차가 있냐를 검정하는 것과 같으며 ,<br>
이때의 귀무가설은 $H_0 : \alpha_1 = ...=\alpha_I$ = 0 으로 표시할 수 있다. <br>
대립가설은 적어도 한 $\alpha$는 0이 아니라고 주어진다.  <br>
모든 처리효과가 같은 때에는 $\sum \alpha_i = 0$ 이라는 가정에 의하여 각각 0이 됨에 유의하자.

* 집단 간 흩어짐은 집단 내 흩어짐에 대한 상대적인 크기로 측정되어야 한다.
    * 집단 내의 흩어짐이 클 때는 집단 간의 차이를 보기가 힘들지만, 집단 내의 흩어짐이 작을 때는 집단 간의 차이가 쉽게 드러나기 때문이다. 

### $(y_{ij} - \bar y)$ = $(y_{ij} - \bar y_i)$ + $(\bar y_i - \bar y)$
(총 제곱합) = (처리 제곱합) + (잔차 제곱합) <br>
=> 여기서 처리제곱합은 '집단간 흩어짐' 을 나타내고, 잔차제곱합은 '집단내흩어짐' 을 나타낸다.

자료의 흩어짐 정도는 제곱합으로 측정될 수 있고, 제곱합을 자유도로 나누면 어떤 미지의 분산의 추정량이 되며 이를 **평균제곱(mean square)**이라고 한다. 
* (평균제곱) = (제곱합) / (자유도) <br>
<br>

#### [ 검정법 정리 ]
#### 검정통계량 : $F = \frac{MSB}{MSW}$ = $\frac{SSB / (k-1)}{SSW / (n-k)}$
#### 검정법 : F > F(k-1, n-k) ; $\alpha$) 이면 $H_0$를 기각한다.

잔차평균제곱 MSW는 항상 $\sigma^2$의 불편추정량이 된다. 또한 귀무가설 하에서는 처리평균제곱 MSB 도 $\sigma^2$ 의 불편추정량이 된다. <br>
그러나 대립가설하에서는 MSB의 기대값이 $\sigma^2$ 보다 크게 되며, <br>
따라서 F비의 값이 클 때에 귀무가설을 기각한다.

## 10.2 이원분류 분산분석
### 10.2.1 임의화 블록 설계
동질성을 가정할 수 없는 경우에는 성질이 유사한 실험단위들끼리 묶어서 '블록'을 만들고, 각 블록 안에서 모든 처리방법들을 랜덤하게 배치하는 실험방법이 타당하다.
### $X_{ij} = \mu + \alpha_i + \beta_j + \epsilon_{ij}$
(관측값 = 전체평균 + 블록효과 + 처리효과 + 오차항)

<br>
TSS = SSB + SST + SSE <br>

* TSS(total sum of squares)는 전체제곱합
* SSB(sum of squares block)는 블록 간 제곱합
* SST(sum of squares treatment) 는 처리방법간 제곱합
* SSE(sum of squares error)는 오차제곱합

=> 각 제곱합들을 해당하는 자유도로 나누어 척도화한 평균제곱합을 구하고, 결과를 요약하면 이원분류 분산분석표를 얻을 수 있다.


### 10.2.2 반복이 있는 이원분류 분산분석
2개의 인자 A,B의 수준수가 각각 p,q이고, 반복수가 r회인 이원배치법을 생각해 보자. <br>
반복이 없는 이원배치법에 비하여 반복이 있는 이원배치법의 가장 큰 장점은 <br>
인자 수준의 조합에서 생기는 효과를 분리하여 구할 수 있다는 것이다. <br>

**[교호작용 interaction]** <br>
인자수준의 조합에서 생기는 효과 <br>
인자 A의 효과가 인자 B의 수준의 변화에 따라 변하는 모형에서 존재한다.

#### 반복이 있는 이원배치법의 모집단 모형
### $Y_{ijk} = \mu + \alpha_i + \beta_j + \gamma_{ij} + \epsilon_{ijk}$
* $\mu$ : 총평균
* $\alpha_i$ : A의 i번째 수준의 효과로서 $\sum_{i=1}^p \alpha_i$ = 0
* $\beta_i$ : B의 j번째 수준의 효과로서 $\sum_{j=1}^p \beta_i$ = 0
* $\gamma_{ij}$ : $A_i$ 와 $B_j$ 의 교호작용 효과로서 $\sum_{i=1}^p \gamma_{ij}$ = $\sum_{j=1}^p \gamma_{ij}$ = 0
* $\epsilon_{ijk}$ : 오차항으로서 서로 독립인 $N(0,\sigma^2)$ 확률변수

반복이 있는 이원배치법에서 관심이 있는 가설은 인자 A,B의 각 수준에서의 효과가 같은가 하는 것 이외에<br>
교호작용의 효과가 확실히 존재하는가 하는 것이다. <br>
교호작용에 대한 귀무가설은 다음과 같다 .<br>
$H_0 : \gamma_{ij} = 0$

TSS = SSA + SSB + SSAB + SSE

#### [Q] : 
만약, 요인 A의 효과에 대한 검정에서는 기각이 되지 않고, 요인B의 효과는 기각이 되면서, A와B의 상호작용은 기각이 될 경우 <br>
요인 A의 효과가 유의하지 않은 것으로 보아 모형에서 제거해야 하는가 ? <br>

#### [A] : 
요인 A의 효과가 상호작용 $\gamma{ij}$에 의해 표현되므로 그 효과가 모형에서 완전히 제거되지 않는다. <br>
따라서, 상호작용이 유의할 때는 요인 A와 B를 모형에 함께 고려하는 것이 통계적으로 의미가 있다.<br>
