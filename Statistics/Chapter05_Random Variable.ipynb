## 5.1 확률변수
확률변수는 확률실험의 결과를 수치적으로 나타내는 함수로서, 표본공간의 원소 각각을 실수값에 대응시키게 된다. <br>
나아가 그 값이 랜덤한 현상의 결과에 따라 달라지므로 '변수'이고, 가질 수 있는 값들이 일어날 가능성이 확률분포로써 표현되는데 <br>
어떠한 확률변수이든지 항상 그에 대응하는 확률분포를 생각할 수 있다.

* 이산형 : 이항분포, 포아송 분포, 기하분포
* 연속형 : 균일분포, 지수분포, 정규분포

### 5.1.1 이산형 확률변수
이산형 확률변수는 유한, 또는 셀 수 있는 무한의 값만을 가질 수 있다.
* 확률변수 X가 x라는 값을 가질 확률 $P_x =P(X=x)$ 는 **확률질량함수**라고 하며, 0과 1 사이의 값을 가지고, 모든 가능한 x값에 대해 더하면 1이 되어야 한다.

### 5.1.2 연속형 확률변수
연속형 확률변수는 어떤 구간 안의 모든 값을 다 취할 수 있는 변수이다. <br>
구간 안의 모든 값은 셀 수 없이 무한하므로, 이산형과 같이 모든 가능한 값 각각에 대해 확률을 대응시킬 수 없다. <br>
따라서 **확률밀도함수** 라고 하는 함수 $f(x)$ 를 이용하여 분포를 표현한다. <br>

* 연속형 확률변수가 어떤 구간에 속하는 값을 가질 확률은 확률밀도함수를 그 구간에서 적분한 값, 즉 밀도함수 아래의 면적과 같다.
* 연속형 확률변수가 단 하나의 값을 가질 확률, 즉 어떤 상수 c에 대하여 $P(X=c)$ 는 0이다.

### 5.1.3 기댓값과 분산
확률변수 X의 기댓값은 일종의 가중평균으로 해석될 수 있으며, 여기서 확률이 가중값의 역할을 하므로 높은 확률을 갖는 결과치가 큰 비중을 차지하게 된다. <br>
**기댓값**은 또한 랜덤한 현상이 독립적으로 여러 번 반복될 때 결과값의 평균 값으로 $E(X)$ 또는 $\mu$ 로 표기한다. <br>
이산형 확률변수 X의 기댓값은 <br>

### $E(X) = \sum_{x} xP(X=x) $ <br>

즉, 확률변수가 가질 수 있는 가능한 값 x에 대하여 가중값의 역할을 하는 확률을 곱한 다음 모든 가능한 값에 대해 더하면 된다.

**# 분산** <br>
확률변수 X의 흐트러진 정도에 대한 측도로 사용되는 분산(variance)은 흔히 $Var(X)$ 또는 $\sigma^2$ 으로 표시하며, <br>
평균으로부터의 편차의 제곱인 $g(X) = (X-E(X))^2$ 에 대한 기댓값, <br>
즉 $Var(X) = E(X-E(X))^2$ 로 정의한다.

* 이 값이 크면 확률변수 X의 흐트러진 정도가 큰 것이고 작으면 흐트러진 정도가 작은 것이다.

**# 표준편차** <br>
표준편차(standard deviation)는 $SD(X)$ 또는 $\sigma$ 로 표기하는데 확률변수 X와 같은 측정단위를 갖는다.
* 예를들어, 변수 X의 단위가 cm이면 $Var(X)$ 의 단위는 $cm^2$ 인 반면에 표준편차는 변수 X와 같은 단위인 cm를 사용하여 흐트러짐을 측정한다.

## 5.2 이산형 확률분포
=> 이산형 확률변수의 대표적인 확률분포로 이항분포, 포아송 분포 설명

### 5.2.1 이항분포
어떤 확률실험이 두 개의 결과만을 가질 수 있을 때 이러한 시행을 베르누이 시행이라고 한다. <br>
두 개의 결과 가운데 하나를 성공 ,다른 하나를 실패라고 하고, 성공이 나올 확률을 p라고 했을때 X의 확률분포는 다음과 같다. <br>
$P_x =P(X=0) = 1-p$ <br>
$P_x =P(X=1) = p$ <br>

* 이러한 베르누이 확률변수 X의 기댓값은 $E(X) = p$ 이고, 분산은 $Var(X) = p(1-p)$ 가 된다.
* 이 베르누이 시행을 정해진 횟수 n번 만큼 독립적으로 반복한다고 했을 때 매 시행에서 성공률 p는 동일하다. 
* n회의 독립반복시행에서 발생하는 성공횟수를 Y라고 한다면 확률변수 Y의 분포는 시행횟수가 n 이고 성공률이 p인 이항분포가 되며, $B(n,p)$ 로 표기한다.

=> 이항분포 $B(n,p)$를 따르는 확률변수 X의 기댓값은 $E(X)=np$ 이고, 분산은 $Var(X) = p(1-p)$ 가 된다.

### 5.2.2 포아송분포
단위시간이나 단위공간에서 일어나는 특성 사건의 횟수에 대한 확률모형을 세울 때 사용되는 분포로서 포아송(Poisson)분포가 있다. <br>
    ex) 어떤 사거리에서 주어진 시간 동안 지나가는 자동차의 대수, 페이지당 나타나는 오타의 숫자 
* 또한, 포아송분포는 이항분포의 시행횟수가 크고 성공률이 아주 낮을 때 이항분포에 대한 근사분포로 사용되기도 한다. <br>

포아송 분포의 확률질량 함수는 다음과 같다. <br>
### $P(X=x) = \frac{exp(-\lambda)\lambda^x}{x!}$

여기서 $\lambda$는 확률변수 X의 기댓값이고, 포아송 분포에서는 기댓값과 분산이 동일하므로 분산 또한 $\lambda$가 된다.  <br>
즉, $E(X) = \lambda$ 그리고 $Var(X) = \lambda$ 이다.

### 5.2.3 그 외의 이산형 분포
#### 기하분포
'성공' 확률이 p인 베르누이 시행을 독립적으로 반복할 때, 첫 번째 '성공'이 일어날 때가지의 시행횟수를 X라고 하면 확률변수 X의 확률질량함수는 <br>
$f(x) = p(1-p)^{x-1}$ <br>
기하 확률변수 X의 기댓값과 분산은 각각 $E(X) = \frac{1}{p}$ , $Var(X) = \frac{1-p}{p^2}$

#### 초기하분포
어떤 주머니에 r개의 빨간 공과 w개의 하얀공(단, r+w = N)이 들어있다. 이제, n개의 공을 무작위 비복원추출 하였을 때,<br>
X를 빨간 공의 개수라고 하면 확률변수 X의 확률질량함수는
$f(x) = \frac{(r,x)(N-r, n-x)}{N, n}$ , x= max(0, n-N+r), ..., min(r,n) <br>

* $N(=r+w)$ 과 $r$이 충분히 크고 $\frac{r}{N}$이 0이나 1에 아주 가깝지 않은 경우ㅜ, 초기하분포의 근사에 이항분포를 활용할 수 있다.

## 5.3 연속형 확률분포
연속형 확률변수의 대표적인 확률분포로 균일분포, 지수분포, 정규분포 설명

### 5.3.1 균일분포
컴퓨터 모의실험으로 0과 1사이에서 난수를 생성하는 경우, 생성된 난수는 0과 1사이의 모든 값을 취할 수 있고 <br>
그 값들은 모두 동일한 확률로 나타난다고 간주될 수 있다. 생성된 난수를 X라고 하면, 이는 구간 (0,1)에서의 균일분표(uniform distribution)를 따른다고 한다. <br>
균일 분포의 경우 확률밀도함수를 주어진 구간에 대해 적분하는 것이 직사각형의 면적을 구하는 문제가 되므로, 균일분포에 대한 확률은 매우 간단히 구할 수 있다.

### 5.3.2 지수분포
포아송 확률과정을 고려했을 때, 특정 사건이 일어난 후 또 다시 같은 사건이 일어날 때까지 걸리는 시간은 음이 아닌 값을 가지는 연속형 확률변수가 된다. <br>
지수분포는 이와 같이 포아송 확률과정에서 도착간격시간을 모형화하거나 생존분포의 일부분을 모형화하는 등에 많이 사용된다. <br>

### $f(x) = \frac{1}{\lambda} exp(\frac{-x}{\lambda})$ , $\lambda > 0$

=> 이때 기댓값과 분산은 각각 $\lambda$ , $\lambda^2$ 이다. 

### 5.3.3 정규분포
시행횟수가 큰 이항분포의 확률을 계산할 방법을 찾다가 이항분포의 근사분포로서 처음 소개되었다. <br>
그 이후 관측에 따른 오차가 정규분포를 따른다는 사실이 밝혀졌으며, 이 분포는 관측값에서 참값을 추정해내는 근본적인 원리로 자리잡게 되었다. <br>

확률밀도함수에서 보이는 것 처럼 정규분포는 두 개의 모수, $\mu$ 와 $\lambda$를 갖는데 $\mu$는 기댓값으로서 분포의 중심을 나타내고,
$\lambda$는 표준편차로서 분포의 퍼짐 정도를 나타낸다. <br>
$\lambda$가 작은 모집단의 분포는 평균 주위에 가깝게 몰려있게 되고, $\lambda$가 큰 모집단의 분포는 넓게 퍼져 있는 형태를 취한다. <br>

#### z-점수(z-score) 또는 표준점수
### $z = \frac{x-\mu}{\sigma}$  <br>
이 값은정규분포에서 추출된 자료 x가 그 평균으로부터 몇 표준편차나 떨어져 있는지를 나타내준다.  <br>
z점수가 -2라면 이 자료는 평균보다 2 표준편차 작은 값이고, z 점수가 3이라면 이 자료는 평균보다 3 표준편차 큰 값이다. <br>

정규분포는 표준편차와 관련하여 다음과 같은 성질을 가진다. 이 성질을 때로 68-95-99.7 규칙이라고도 한다. <br>
관측값의 68%는 평균에서 1 표준편차 이내에 들어간다. $\mu + \sigma$ <br>
관측값의 95%는 평균에서 2 표준편차 이내에 들어간다. $\mu + 2\sigma$ <br>
관측값의 99.7%는 평균에서 3 표준편차 이내에 들어간다. $\mu + 3\sigma$ <br>

# 5.4 결합분포
여러 개의 확률변수들을 동시에 고려할 때, 이들의 결합분포(joint disstribution)에 대한 이론이 필요하게 된다. <br>
ex) 아버지의 키와 아들의 키를 짝을 이루어 함께 생각한다던가, 환자에 대한 건강기록을 구성하는 변수로 나이, 혈압, 몸무게 등을 생각한다던가 <br>

### 5.4.1 결합 확률질량함수 및 결합 확률밀도함수
이산형 확률변수 X와 Y의 **결합 확률질량함수**(joint probability mass function) 는 모든 가능한 (x,y)쌍에 대해 <br>
$p(x,y) = P(X=x$ 이고 $Y=y)$ 로 정의된다. <br>

연속형 확률변수 X,Y에 대해서는 **결합 확률밀도함수**(joint probability density function)로 그 결합확률분포를 나타낸다.<br>

### 5.4.2 조건부분포
여러 개의 확률변수들 중 몇 개의 값이 주어졌다는 조건하에서 나머지 변수들의 분포를 구하고자 할 때가 있다. <br>
가령, 아버지의 IQ와 성인 자녀의 IQ가 짝지어져 있을 때, 아버지의 IQ가 120인 자녀들의 IQ는 어떤 분포를 따를까 ?  라는 질문에 '조건부분포'를 통해 파악될 수 있다.<br>

### 5.4.3 기댓값과 공분산
두 확률 변수 X와 Y의 연관관계의 척도인 **공분산**(covariance)은 $Cov(X,Y)$ 로 표기하는데, 이는 X와 Y의 함수인 h(X,Y) = (X-EX)(Y-EY) 의 기댓값으로
$Cov(X,Y) = E((X-E(X)))(Y-E(Y)) = E(XY) - E(X)E(Y)$ 와 같이 정의된다.
* 공분산은 X와 Y사이의 연관관계를 측정하는 척도로서 단위에 영향을 받는다.
    * 즉, 단위를 바꾸면 공분산의 값도 바뀌게되며, 단위가 다른 경우의 공분산을 직접 비교하기 어렵다.
    * 이러한 단점을 보완하기 위해 단위에 영향을 받지 않는 **상관계수**를 다음과 같이 정의하여 사용한다. <br>
### $\rho(X,Y) = Corr(X,Y) = \frac{Cov(X,Y)}{\sqrt{Var(X)Var(Y)}}$

한편, 두 확률변수 X,Y가 서로 독립이면 $E(XY) = E(X)E(Y)$이므로, 공분산과 상관계수 모두 0이다. <br>
그러나 *상관계수가 0이라고 해서 두 변수 X와 Y가 독립인 것은 아니며*, *상관계수=0인 X와 Y는 상관관계가 없다*라고 말한다. <br>
상관계수는 항상 -1에서 1사이의 값을 가진다.
