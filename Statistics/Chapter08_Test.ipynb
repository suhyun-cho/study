## 8.1 가설검정의 기본원리
### 8.1.1 가설
가설은 물론 모집단의 성질에 대한 것이고, 보통 모수의 값에 대한 가정이다. 두 종류의 가설 중 하나는 *귀무가설* 또는 영가설이라 하고, 다른 하나는 *대립가설* 이라고 한다. <br>
일반적으로 귀무가설은 기존의 사실, 주장, 믿음 등으로 해석되고, 대립가설은 새로운 사실, 주장, 믿음 등으로 해석된다. <br>
보통 연구자들이 통계적 검정을 하게 되는 이유는 기존의 사실과 달라진 것이 있다는 주장을 밝히기 위한 것이므로, 대립가설에 주어지는 모수영역이 연구자가 입증하고자 하는 영역이 되고 귀무가설에 주어지는 모수영역은 그에 반대되는 영역이 된다.

### 8.1.2 검정통계량
모수를 추정하기 위해서 **추정량** 이라고 하는 통계량을 사용하는 것처럼, 모수를 검정하기 위해서는 **검정통계량** 이라고 하는 통계량을 사용한다. <br>

<예제 8.2>
표본평균이 $\bar X_n$ 이고 귀무가설에 제시된 모평균이 $\mu_0$ 라면, 검정통계량은 $\frac{\bar X_n - \mu_0}{\sigma / \sqrt n}$으로 잡을 수 있다. <br>
검정통계량은 '통계량' 이므로 그 자체가 확률변수이고, 미지의 모수가 포함되어서는 안 된다. <br>
$\frac{\bar X_n - \mu_0}{\sigma / \sqrt n}$ 은 표본평균이 귀무가설로부터 몇 표준편차만큼 떨어져있는지를 측정해 주고 있으므로,<br>
이 값이 아주 작다면 귀무가설을 기각하고 대립가설을 뒷받침하는 증거가 있다고 결론을 내릴 수 있다.

### 8.1.3 유의확률(P-value)
* p-value 정의 : 귀무가설하에서 검정통계량의 관측값이 귀무가설을 기각하는 방향으로 나타나는 확률

귀무가설이 사실이라는 가정하에서, 검정통계량이 관측값과 같거나 혹은 이보다 멀리 떨어졌다는 것은 귀무가설에 대한 반증이 더 강한, 그래서 대립가설을 더 잘 뒷받침해줄 수 있는 관측값을 의미한다. <br>
따라서 유의확률이 작을수록 귀무가설에 대해 더 강한 반증이라고 할 수 있다. <br>

유의확률이 크다면, 귀무가설하에서 관측값과 같거나 더 극단적인, 즉 귀무가설로부터 더 멀리 떨어져 있는 자료를 얻을 확률이 높다는 것이므로 주어진 관측값이 귀무가설에 대한 반증이 되지 못한다.<br>
따라서, 유의확률이 클 때는 귀무가설을 기각하지 못한다.<br>

반대로 유의확률이 작다면 귀무가설하에서 관측값과 같거나 더 극단적인 자료를 얻을 확률이 낮다는 것이므로 관측값이 희귀한 값이라는 것이다. 
이렇게 희귀한 값이 관측되었으므로, 이 관측값은 귀무가설이 참이 아니라는 증거가 될 수 있다. 그러므로 유의확률이 작을 때는 귀무가설을 기각한다. <br>

유의 확률이 크고 작음을 판단하는 기준으로는 '유의수준'이 흔히 사용된다.

### 8.1.4 두 가지 오류와 유의수준 
가설검정에서 발생할 수 있는 오류는 두 가지가 있는데, <br>
귀무가설이 참일 때 이를 기각하는 오류는 **제1종오류** 라고 하고, 귀무가설이 거짓일 때 귀무가설을 기각하지 못하는 오류를 **제2종오류**라고 한다.<br>

형사재판의 예에서는, 실제로는 무죄인 피고인에게 유죄가 선고되어 죄 없는 사람이 감옥에 가게 되는 것이 제1종 오류, <br>
실제 유죄인 피고인에게 무죄가 선고되어 범죄자가 풀려나게 되는 것이 제2종오류이다.<br>

제1종 오류를 저지를 확률의 최댓값을 검정에 있어서 유의수준(significance level)이라고 하고 이를 통상 $\alpha$로 표시한다. 
제2종 오류를 범할 확률은 보통 $\beta$를 써서 표시하는데, 이를 1에서 뺀 값을 검정의 검정력(power)라고 한다. <br>
즉, **검정력**은 귀무가설이 참이 아닐 때, 옳은 결정을 내려 귀무가설을 기각하게 되는 확률을 말한다. <br>

*유의수준* 은 귀무가설이 참이라는 가정하에 계산되므로 검정통계량의 귀무가설하에서의 분포를, <br>
*검정력* 은 대립가설이 참이라는 가정하에 계산되므로 검정통계량의 대립가설하에서의 분포를 이용하여 계산한다. <br>

제1종 오류와 제2종 오류는 이율배반적이라서 두 종류의 오류를 범할 확률을 동시에 최소화하는 검정법을 고를 수는 없다.

귀무가설을 기각하는 경우, 결과가 "통계적으로 유의하다"라는 표현을 쓰곤 한다. <br>
통계적으로 유의하다는 것은 표본에서 얻어진 결과가 우연히 발생한 것이 아니라는 뜻이다. 그런데 통계적인 유의성이 반드시 실제적인 유의성과 연결되지는 않는다. <br>
통계적 유의성은 어떤 효과가 있다는 것만 말해줄 뿐, 그 효과의 크기에 대해서는 말해주지 않는다. <br>

=> 차이가 있음이 매우 유의하다는 것이 차이가 크다는 뜻은 아니기 때문이다.

## 8.2 모평균에 대한 검정
모평균이 $\mu$ 이고 분산이 $\sigma^2$ 인 정규분포에서 뽑힌 크기가 $n$인 랜덤표본 $X_1, X_2,,,X_n$ 으로부터 모평균이 $\mu_0$ 와 같은지 그렇지 않은지를 검정해보자. <br>
이때 귀무가설은 $H_0:\mu=\mu_0$ 로 할 수 있고, 대립가설은 $H_1: \mu > \mu_0, H_1: \mu < \mu_0$, 또는 $H_1:\mu \neq \mu_0$ 의 세가지로 생각할 수 있다. <br>

대립가설이 $\mu < \mu_0$ 또는 $\mu > \mu_0$ 가 되는 경우를 **단측검정**(one-sided test) 라고 하고,<br>
대립가설이 $\mu \neq \mu_0$ 가 되는 경우를 **양측검정**(two-sided test)이라고 한다. 

#### [모분산이 알려져 있는 경우]
모분산이 알려져 있다면 $Z = \frac{\bar X - \mu_0}{\sigma / \sqrt n}$ 는 미지의 모수가 포함되어 있지 않은 통계량이 되고, 이를 검정통계량으로 사용할 수 있다.<br>
이 통계량은 표본평균이 귀무가설에서 가정된 값 $\mu_0$에서 몇 표준편차만큼 떨어져 있는지 그 거리를 측정해 준다. <br>

대립가설이 $\mu > \mu_0$ 인 경우에는 검정통계량 Z의 관측값이 충분히 클 때 , 귀무가설을 기각하고 <br>
대립가설이 $\mu < \mu_0$ 인 경우에는 검정통계량 Z의 관측값이 충분히 작을 때, 귀무가설을 기각한다. <br>
 $\mu \neq \mu_0$ 이면 검정통계량 Z의 관측값이 충분히 클 때 또는 작을 때 귀무가설을 기각한다.

#### [모분산이 알려져 있지 않은 경우]
모분산이 알려져 있지 않으면, $Z = \frac{\bar X - \mu_0}{\sigma / \sqrt n}$ 에 미지의 표준편차 $\sigma$ 가 포함되어, 이를 검정통계량으로 사용할 수 없다. <br>
7장에서 $\mu$의 참값을 추정할 때 했던 것처럼 $\sigma$를 표본 표준편차인 S로 대치한 $t = \frac{\bar X - \mu_0}{S/\sqrt n}$를 검정통계량으로 고려하면, <br>
이는 귀무가설 $H_0 : \mu = \mu_0$ 가 참이라는 가정하에서 자유도가 n-1인 t분포를 따르게 된다. <br>
이것을 검정통계량으로 사용한 모평균에 대한 검정을 **일표본 t-검정** 이라고 한다.

## 8.3 모비율에 대한 검정
성공과 실패의 두 값만을 가질 수 있는 변수의 모집단에서의 성공률을 $p$ 라고 하고, 표본에서의 성공률을 $\hat p$이라고 하자. <br>
중심극한 정리에 의해 표본의 크기가 클 때 표본비율 $\hat p$은 평균이 $p$이고 분산이 $\frac{p(1-p)}{n}$ 인 정규분포를 근사적으로 따른다는 것을 이용하여 모비율에 대한 검정을 할 수 있다. <br>

귀무가설은 $H_0 :p = p_0$ 으로 놓고 대립가설은 $H_1:p>p_0, p< p_0$ 또는 $p \neq p_0$ 으로 놓는다. <br>
$\hat p$는 근사적으로 $N (p, \frac{p(1-p)}{n})$ 분포를 따르므로 귀무가설하에서 $Z = \frac{\hat p - p_0}{\sqrt \frac{p_0()1-p_0}{n}}$ 는 근사적으로 표준정규분포를 따르게 된다. <br>

이 통계량은 표본비율이 귀무가설에서 가정된 값에서 얼마나 떨어져 있는지를 측정해 주는데, <br>
가령 Z=2라면 이는 관측된 표본비율이 귀무가설하에서 $p_0$ 보다 2 표준편차만큼 더 큰 값임을 의미한다.

## 8.4 모평균의 차에 대한 검정
두 독립인 표본을 통해 모평균의 차에 대한 검정을 하는 경우, <br>
두 분산이 알려져 있으면 정규분포를 따르는 Z-통계량을 검정통계량으로 쓰는 Z-검정을 하고 <br>
### $Z=\frac{\bar X_n - \bar Y_m}{\sqrt{\frac{\sigma_X^2}{n} + \frac{\sigma_Y^2}{m}}}$

두 분산이 알려져 있지 않으면 t-검정을 한다. <br>
t-검정에서 먼저 두 모분산이 같다고 가정할 수 있는지 확인하고, 같다고 가정할 수 있으면 '합동분산추정량'을 이용하여 검정하고, <br>
### $t = \frac{\bar X_n - \bar Y_m}{S_p \sqrt{\frac{1}{n} + \frac{1}{m}}}$

같다고 가정할 수 없는 경우 각 모분산을 표본분산으로 대체하여 근사적인 t-검정을 한다 .
### $t = \frac{\bar X_n - \bar Y_m}{\sqrt{\frac{S_X^2}{n} + \frac{S_Y^2}{m}}}$

두 모분산이 같다고 가정할 수 있는지 확인하기 위해서는 F-검정을 쓸 수 있다. <br>
모분산이 같다고 가정하기 어려운 경우에는 같지 않다고 가정하고 근사 t-검정을 쓰면 된다.

## 8.5 모비율의 차에 대한 검정
예) 어떤 여론조사기관이 2017년 1월 12일 발표한 1월 9~11일 조사결과에서, 여당인 새누리당의 지지도는 서울 지역에서 14.9%, 대전/충청/세종 지역에서 13.7%를 기록했다. <br>
이때 서울 지역에서의 지지도가 충청 지역에서의 지지도보다 높다고 할 수 있을까? <br>
14.9%가 13.7%보다 크긴 하지만, 표본오차를 생각할 때 서울 지역의 모비율이 충청 지역의 모비율보다 반드시 크다고 말할 수는 없을 것이다. <br>
이런 경우 모비율의 차에 대한 검정이 필요 <br>

귀무가설 $H_0:p_X = p_Y$ 하에서는 $p_X = p_Y$ 이므로 이를 P로 표기하면, <br>
### $\frac{(\hat p_X - \hat p_Y)-0}{\sqrt{p(1-p)(\frac{1}{n}+\frac{1}{m})}}$
은 귀무가설하에서 근사적으로 표준정규분포를 따르게 된다. <br>

하지만 귀무가설이 맞다고 해도 여전히 $p=p_X=p_Y$는 미지의 값이다 .귀무가설하에서는 $p_X$ 가 $p_Y$와 같다는 것을 알 뿐이지 <br>
그 값이 무엇인지는 가정되지 않기 때문이다 .그러므로 위의 검정통계량은 사용될 수 없다. <br>

따라서 p를 추정해야 하는데, 귀무가설하에서는 두 모비율이 같으므로($p_X = p_Y$) , 두 표본을 통합하여 합동추정량 을 쓰면된다 .<br>
### $\hat p = \frac{n\hat p_X + m\hat p_Y}{n+m}$ 
을 계산하여 <br>
### $\frac{(\hat p_X - \hat p_Y)-0}{\sqrt{\hat p(1-\hat p)(\frac{1}{n}+\frac{1}{m})}}$ 
을 검정통계량으로 사용한다.<br>

## 8.6 모분산에 대한 검정
### 8.6.1 단일모분산에 대한 검정
표본분산이 $S^2$ 이고 표본평균이 $\bar X$일때, $(n-1)S^2/\sigma_0^2$ 은 귀무가설하에서 자유도가 n-1인 카이제곱분포를 따르므로, 이를 검정통계량으로 사용할 수 있다. <br>

모평균 $\mu$가 알려져 있으면 모분산검정에 대한 검정통계량은 
### $\frac{\sum_{i=1}^n (X_i - \mu)^2}{\sigma_0^2}$ 
을 사용할 수 있는데, 이는 귀무가설 $H_0 : \sigma^2 = \sigma_0^2$ 하에서 자유도가 n인 카이제곱분포를 따른다. <br>

따라서 자유도가 n인 카이제곱분포를 사용하는 것 외에는 모평균 $\mu$가 알려져 있지 않은 경우와 동일한 과정을 거쳐 유의확률을 구할 수 있다.

### 8.6.2 모분산의 비에 대한 검정
두 정규모집단의 모분산 동일성에 대한 검정은 모분포의 흐트러진 정도를 비교하는 목적으로 사용된다. <br>
가령, 두 가지 교육방법의 성적분포의 흐트러진 정도를 비교하기 원할 때, <br>
또는 두 집단의 평균을 비교하는 과정에서 '합동분산추정량'을 사용할 것인지 아닌지를 결정하기 위해 모분산의 동일성을 검정할 수도 있음. <br>

두 모분산의 동일성여부에 대해서 귀무가설은 $H_0:\sigma_X^2 = \sigma_Y^2$ 이라고 해놓고,
### $F = \frac{S_X^2 / \sigma_X^2}{S_Y^2 / \sigma_Y^2}$
이 자유도가 (n-1, m-1)인 F분포를 따르며, 귀무가설 $H_0$ 하에서는 $\sigma_X^2 = \sigma_Y ^2$ 이므로 
### $F=S_X^2 / S_Y^2$
이 된다.

한편, 모평균 $\mu_X$ 와 $\mu_Y$가 알려져 있으면 모분산의 동일성검정에 대한 검정통계량은 
### $F = \frac{\sum_{i=1}^n (X_i-\mu_X)^2 / n}{\sum_{j=1}^m (Y_j-\mu_Y)^2 /m}$
이 되는데 유위확률 계산에 자유도가 (n,m)인 F분포를 사용하는 것 외에는 모평균 $\mu_X$ 와 $\mu_Y$가 알려져 있지 않은 경우와 동일한 과정을 거친다 .
