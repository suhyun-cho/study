그 모수의 참값이 무엇인지 자료를 통해 합리적인 추측을 하는 것이 **추정**이고, <br>

모수의 추정에 사용되는 통계량을 추정량(estimator)이라고 하고, <br>
검정에 사용되는 통계량을 검정통계량(test statistic)이라고 한다. <br>

## 7.1 점추정과 구간추정
* 점 추정
    * 하나의 값으로 모수를 추정하는 것
    * 추정값의 정확성에 대한 정보를 제공하지 않는다. <br>
    ex) 우리나라 고등학생의 평균키는 170cm 다 <br>
    
* 구간 추정 
    * 적절한 구간을 설정하여 모수를 추정하는 것
    * 모수가 어떤 값을 가질 것인지에 대한 정보와 함께 신뢰도와 오차의 한계를 통하여 추정의 불확실성에 대한 정보도 제공
    (점 추정과의 차이) <br>
    ex) 우리나라 고등학생 평균키는 167cm 에서 173cm 사이다.
    * 추정값의 정확성에 대한 정보를 제공하지 않는다.
* 구간 추정 
    * 적절한 구간을 설정하여 모수를 추정하는 것
    * 모수가 어떤 값을 가질 것인지에 대한 정보와 함께 신뢰도와 오차의 한계를 통하여 추정의 불확실성에 대한 정보도 제공
    (점 추정과의 차이)

### 7.1.1 점추정
모평균을 추정할 때 사용될 수 있는 표본평균, 표본중위수, 표본절사평균 등을 추정량(estimator)이라고 한다. <br>
* 실제로 표본을 관측하여 이들의 값을 계산하면 그 값들은 추정값(estimate)이 됨.

추정량의 좋고 나쁨을 관측값 하나로 판단하는 것음 무리다. (궁수가 한번 잘쐈다고 훌륭하다고 볼 수 없는것과 동일) <br> 
표본을 많이 뽑아서 각 표본에 대한 추정값을 계산하여 그 값들의 분포를 보고, 즉 추정량의 분포를 통해 결정해야 한다.  <br>

----->  Q : 왜 추정량의 분포를 고려하여 추정해야 하나 ? <br>
----->  A : 추정량은 통계량이고 통계량은 확률변수이므로 표본이 달라지면 값이 달라지기 때문<br>

#### **[ 편향과 분산 ]**
#### [ 편향과 분산 ]
* 분산
추정값들이 가까이 있으면 추정량의 분산이 작은 것이고, <br>
넓게 퍼져 있으면 분산이 큰 것이다. <br>

* 편향
추정값들이 모수의 참값을 중심으로 분포하면, 즉 추정량의 기댓값이 모수의 참값과 같으면 추정량은 편향이 없는 것이고 <br>
그렇지 않으면 추정량은 편향이 있는 것이다.<br>
=> 추정량의 편향은 추정량의 기댓값에서 모수의 참값을 뺀 것으로 정의.

<img src="var .png" width="300"> <br>

=> 분산도 작고 편향의 절댓값도 작은 것이 좋은 추정량 <br>
=> 추정의 기준을 평균제곱오차(MSE)로 잡을 수 있다. <br>
    평균제곱오차(mean square error) = 분산 + $편향^2$ <br>
###    MSE = $\frac{1}{n}\sum_{i=1}^n (\hat Y_i - Y_i)^2 $ <br>

* 특별히, 편향이 0인 추정량은 비편향추정량이라고 한다.
    * 추정량의 평균이 실제 모수값과 일치할 때 추정량을 비편향추정량이라고 한다.

#### < 예제 7.1 >
결론 : 표본평균의 평균은 모평균과 같게되어 표본평균은 모평균의 비편향추정량이 된다. <br>

| $\bar X$ |1|1.5|2|2.5|3|3.5|4|4.5|5|
| --- | --- |---|---|---|---|---|---|---|---|
| P($\bar X$)|$\frac{1}{25}$|$\frac{2}{25}$| $\frac{3}{25}$ |$\frac{4}{25}$|$\frac{5}{25}$ |$\frac{4}{25}$|$\frac{3}{25}$|$\frac{2}{25}$|$\frac{1}{25}$| 

* 중간으로 갈수록 확률이 커지고 양 끝으로 갈 수록 확률이 작아짐 
    * 정규분포와 흡사

#### < 예제 7.2 >
결론 : 표본분산의 '평균'은 모분산과 같게 되어 표본분산은 모분산의 비편향추정량이 된다. <br>
* 분산은 모분산 $\sigma^2$ 의 1/n배인 $\sigma^2$/n 과 같다.
    * 여기서 n은 표본의 개수 
    평균제곱오차(mean square error) = 분산 + $편향^2$

* 특별히, 편향이 0인 추정량은 비편향추정량이라고 한다.

### 7.1.2 구간추정
구간추정은 점추정과 달리 추정의 불확실성에 대한 정보도 함께 제공한다. <br>
표본평균으로 모평균을 점추정할 때, 모평균이 표본평균과 정확히 같다고 볼 수 없다. 
* why ? 모평균은 하나의 상수값인데 표본평균은 표본에 따라 다른 값을 취할 수 있기 때문.

모평균이 $\mu$ 이고 모분산이 $\sigma^2$ 인 모집단에서 뽑은 크기 $n$인 표본에 대하여 표본평균 $\bar X_n$ 의 기댓값은 $\mu$ 이고 
표준편차는 $\frac{\sigma}{\sqrt n}$  이다. <br>



즉, 표본의 크기가 커질수록 표준편차가 작아져, 표본평균의 분포가 $\mu$ 주변으로 집중된다는 것이다.<br>

* 그러므로 표본의 크기가 충분히 크면 관측된 표본평균이 모평균에 매우 가까이 있다고 생각 할 수 있다.<br>

* 여기서 표준편차  $\frac{\sigma}{\sqrt n}$  는 추정량인 표본평균 $\bar X_n$ 의 흩어진 정도에 대한 측도로서 **"표준오차"** 라고도 한다.
    * 표준오차 => 표본평균들의 표준오차

-----> Q. : '표준편차' 와 '표준오차' 의 차이점은 ? <br>
-----> A. : 
* 표준편차 : 모집단이건 표본집단이건 그 안의 숫자들의 편차를 뜻함. (평균을 중심으로 퍼진 정도)
* 표준오차 : 여러개 뽑히는 표본평균들의 편차 (뽑힌 표본집단들이 모집단을 추정하기에 적절한지)
    * 즉, 표준오차가 작다면 표본평균들로 모평균을 추정한다는 사실을 의심할 필요가 적어짐.
    * 표본에 대해서만 논하지, '모표준오차'와 같은 개념은 잘못된것.

즉, 표본의 크기가 커질수록 표준편차가 작아져, 표본평균의 분포가 $\mu$ 주변으로 집중된다는 것이다.<br>

그러므로 표본의 크기가 충분히 크면 관측된 표본평균이 모평균에 매우 가까이 있다고 생각 할 수 있다.<br>

여기서 표준편차  $\frac{\sigma}{\sqrt n}$  는 추정량인 표본평균 $\bar X_n$ 의 흩어진 정도에 대한 측도로서 **표준오차** 라고도 한다.<br>
* 표본평균의 표준오차 => 표준오차

#### [ 표본평균의 분포 ]
* 모집단이 정규분포를 따르면, 표본평균의 분포는 $N(\mu, \frac{\sigma^2}{n})$
###    * $P(\bar X_n - z_{\frac{\alpha}{2}} \frac{\sigma}{\sqrt n}  <= \mu <= \bar X_n + z_{\frac{\alpha}{2}} \frac{\sigma}{\sqrt n} )$ 
이 성립한다.
    * 예를들어 $\alpha$ = 0.05 일때 위 식은 표본평균의 95% 정도가 모평균으로부터 (1.96 X 표준편차) 이상 떨어져 있지 않다는 뜻이 된다. 
    * 즉, 매우 많은 표본을 뽑아서 각각에 대해 표본평균을 계산하여 $\bar X_n \pm z_{\frac{\alpha}{2}} \frac{\sigma}{\sqrt n}$ 의 구간을 구하면 그 가운데 95% 정도의 구간이 모평균을 포함하게 된다는 것이다.
    * $\bar X_n \pm z_{\frac{\alpha}{2}} \frac{\sigma}{\sqrt n}$  의 구간은 표본에 따라달라지는 랜덤한 구간으로, 모평균을 포함할 수도 있고 포함하지 않을 수도 있다. 
    
* 모분포가 정규분포가 아니더라도, 표본의 크기가 충분히 커서 중심극한정리를 사용할 수 있다면 표본평균의 분포를  $N(\mu, \frac{\sigma^2}{n})$으로 근사할 수 있음 

#### [ 95% 신뢰수준 또는 신뢰도 의미 ]
* 동일한 크기의 표본을 매우 많이 뽑아 각 표본에 대해 $\bar X_n \pm z_{\frac{\alpha}{2}} \frac{\sigma}{\sqrt n}$ 의 구간들을 구하면 그 중에서 95% 정도가 모평균을 포함한다는 의미
* 이 구간을 모평균에 대한 신뢰구간이라고 하며 이것이 모평균의 구간추정이 된다.

신뢰도 = 신뢰도 수준 (+/-) 표본오차

#### [ 신뢰구간의 정확한 의미 ]
-----> Q. 다음 중 맞는 의미

2. 같은 방법으로 100번 표본을 추출했을 때, 함께 계산되는 100개의 신뢰구간 중 모평균을 포함한 신뢰구간들의 숫자가 95개정도 된다.

### 7.2 모평균의 신뢰구간
모분산이 알려져 있는 경우 <br>
### $\bar X_n \pm z_{\frac{\alpha}{2}} \frac{\sigma}{\sqrt n}$

모분산이 알려져 있지 않은 경우  <br>
### $\bar X_n \pm t_{\frac{\alpha}{2}} \frac{(n-1)S}{\sqrt n}$

<정리>
1. 모분산을 알때 ----> Z분포
2. 모분산이 알려져 있지 않을 때 <br> 
    2-1. 소표본 ---> t분포 <br>
    2-2. 대표본 ---> Z분포 <br>


평균이 $\mu$ 이고 분산이  $\sigma^2$ 인 정규분포에서 크기가 $n$ 인 임의표본을 뽑을 때 $\frac{\bar X_n - \mu}{\frac{\sigma}{\sqrt n}}$ 은 표준정규분포를 따른다. <br>
나아가, 표본평균 $\bar X_n$의 표준오차 $\sigma / \sqrt n$ 를 $S / \sqrt n$ 으로 추정하여 대체한 $\frac{\bar X_n - \mu}{\frac{S}{\sqrt n}}$ 는 자유도가 n-1인 t분포를 따른다. <br>
이러한 사실에 근거하여 모평균 $\mu$에 대한 신뢰수준 1-$\alpha$ 인 신뢰구간 $\bar X_n \pm t_{\frac{\alpha}{2}} \frac{(n-1)S}{\sqrt n}$ 을 구할 수 있다.

### 7.3 모비율의 신뢰구간
* 모비율 : 모집단이 어떤 특성을 가지는 비율
* 표본비율 : 모집단에서 임의로 추출한 표본에서 어떤 특징을 가지는 대상의 비율

성공할 확률을 모비율 $p$라고 하고, 표본에서의 성공률을 표본비율 $\hat{p}$ 라고 하자. <br>
$\hat{p}$  = $\frac{X}{n}$ = (대상수=확률변수 / 표본크기) <br>
$\hat{p}$  = $\frac{X}{n}$ = (대상수:확률변수 / 표본크기) <br>

표본의 크기가 n일때, $n\hat{p}$ 은 표본에서 '성공'이 관측되는 횟수이므로 <br>
성공 확률이 $p$인 베르누이 시행을 $n$번 독립적으로 반복했을 때의 성공횟수로 볼 수 있고, <br>
이는 이항분포 $B(n,p)$ 로 모형화할 수 있다.

이항분포 $B(n,p)$에 n이 크면 정규분포를 따르므로, <br>
$N(p, \frac{pq}{n})$ 

이때 $\hat{p}$ 의 표준오차는 $\sqrt \frac{p(1-p)}{n}$ 이 된다.

모비율 $p$ 를 표본비율 $\hat{p}$ 로 추정하여 신뢰수준 $1-\alpha$인 모비율의 신뢰구간을 <br>
### $\hat{p} \pm z_\frac{\alpha}{2} \sqrt \frac{\hat{p}(1-\hat{p})}{n}$ <br>


#### < 예제 7.6 >
**## 표본오차** <br>
* 모집단을 대표할 수 있는 전형적인 구성 요소를 선택하지 못함으로써 발생하는 오차 (최대허용오차) <br>
    * 모집단 평균과 표본평균간의 차이
    * 쉽게말하면, 표본의 평균에 의해 모집단의 평균을 추정할 때의 신뢰구간을 의미.

**## 표준오차** <br>
* 여러개 뽑히는 표본평균들의 편차 (뽑힌 표본집단들이 모집단을 추정하기에 적절한지)

여기서 '표본오차'는 <br>
### $Z_\frac{\alpha}{2} \sqrt\frac{\hat p (1-\hat p)}{n}$ 이고, <br>

'표준오차' 는 <br>
### $ \sqrt\frac{\hat p (1-\hat p)}{n}$ 이다. <br>
=> 통계학에서 '표본오차'는 일정한 신뢰수준을 확보하기 위한 Z값 또는 t값에 표준오차를 곱한다.

------------------------------------------------------------------------------------------------------------------
$\hat{p} \pm z_\frac{\alpha}{2} \sqrt \frac{\hat{p}(1-\hat{p})}{n}$ 로 구하게 된다.

### 7.4 표본크기의 결정
===> 요약 : 표본크기를 결정하는 이유는, 적절한 신뢰구간과 신뢰수준을 얻기 위함. 

신뢰구간을 구할 때, $\bar X_n$ , $\sigma$ , $n$ 이 모두 고정된 상태에서 신뢰수준을 높이면, 즉 $\alpha$를 줄이면 $z_\frac{\alpha}{2}$ 의 값이 커져서 구간이 넓어진다. <br>
(ex. $\alpha$ = 0.1 이면 $z_\frac{\alpha}{2}$=1.645 이지만 $\alpha$=0.05이면 $z_\frac{\alpha}{2}$=1.96 이 된다. )

* 신뢰수준을 높이기 위해 너무 넓은 신뢰구간을 제공하면 그 추정은 별 의미가 없을 수 있다. <br>
ex) 성인남자의 키에 대한 추정에서 평균키에 대한 신뢰구간이 150cm 에서 200cm라고 할 때 <br>

* 반대로 신뢰수준이 너무 낮아도 의미가 없음
ex) 성인남자의 평균키에 대한 신뢰구간이 170cm에서 171cm 라고 할때, 신뢰수준이 50%라면 이 추정을 신뢰하기 어려움. <br>

신뢰구간의 길이가 $d$ 라고 했을때, <br>
#### $d$ = 2 x $z_\frac{\alpha}{2} \sigma / \sqrt n$ 으로부터 <br>
#### n = $(\frac{2 z_\frac{\alpha}{2} \sigma}{d})^2$ 을 얻을 수 있다. <br>

표본크기를 결정하는 데 있어서 비용이나 다른 제약조건을 고려해야 하므로, 가장 작은 표본크기를 결정하는 것이 좋다. <br>
n >= ($\frac{2 z_\frac{\alpha}{2} \sigma}{d}$)^2  을 만족하는 가장 작은 표본크기를 결정하면 된다. <br>


* 모비율에 대한 신뢰구간을 필요로하는 연구에서는  <br>
모비율에 대한 신뢰구간을 필요로하는 연구에서는  <br>
2 x $z_\frac{\alpha}{2} \sqrt \frac{p(1-p)}{n}$  로부터 표본의 크기를 결정해야 하는데, p는 모르는 모수이고 표본을 얻기 전이므로 p를 $\hat{p}$ 로 대체할 수 없는 문제가 있다. <br>
p(1-p) = 1/4 로 놓고 $d = \frac{2 z_\frac{\alpha}{2}}{\sqrt 4 n}$ = $\frac{z_\frac{\alpha}{2}}{\sqrt n}$ 를 풀어서 <br>
$n = (\frac{z_\frac{\alpha}{2}}{d})^2$ 으로 표본의 크기를 결정할 수 있다. 

* $p(1-p)$ = 1/4로 놓는 것은 p = 1/2 에서 최댓값을 가지며, 최댓값을 쓰게 되면 실제로 필요한 표본의 크기보다 더 큰 표본을 쓰게 되어 보수적인 신뢰구간을 얻기 때문이다.

### 7.5 모평균의 차에 대한 신뢰구간
참고자료 : http://blog.naver.com/PostView.nhn?blogId=lchry&logNo=220509297420 

두 평균의 차이에 대한 추정이나 검정에 있어서 표본평균의 차이뿐만 아니라, 표본 내의 산포 또한 고려해야 함. <br>
즉, 표본 내의 산포에 비하여 표본평균의 차이가 클 때 두 모평균에 차이가 있다고 말한다. <br>

두 모평균의 차이 $\mu_X - \mu_Y$ 에 해당하는 표본평균의 차이 $\bar X_n - \bar Y_n$ 는  <br>
#### $N(\mu_X - \mu_Y , \frac{\sigma^2_X}{n} +\frac{\sigma^2_Y}{n} )$ 
분포를 따른다.<br>

* **두 모분산이 알려져 있는 경우**
    * 신뢰구간을 $\bar X_n - \bar Y_m \pm Z\frac{\alpha}{2} (\frac{\sigma^2_X}{n} + \frac{\sigma^2_Y}{m})^\frac{1}{2}$ 와 같이 얻을 수 있다.
    * 이때의 검정통계량은
###    $Z = \frac{\bar X_n - \bar Y_m -(\mu_X - \mu_Y)}{(\frac{\sigma^2_X}{n}+\frac{\sigma^2_Y}{m})^\frac{1}{2}}$ <br>


* **두 모분산이 알려져 있지 않은 경우**
    * 1) 두 모분산이 같다고 가정할 수 있는 경우 (등분산 가정 충족, F검정)
        * 모분산을 모르더라도 두 모분산 $\sigma^2_X$ 과 $\sigma^2_Y$ 이 같다고 가정할 수 있는 경우에는 '합동이표본분석'을 할 수 있다. 
            * 이때의 자유도는 n+m-2 이고, 합동분산추정량 $S_p$ 는 
####            $S^2_p = \frac{(n-1)S^2_X + (m-1)S^2_Y}{n+m-2}$
### t = $\frac{(\bar X_n - \bar Y_n) - (\mu_X - \mu_Y)}{S_p (\frac{1}{n} + \frac{1}{m})^\frac{1}{2}}$
   

   * 2) 두 모분산이 같다고 가정할 수 없는 경우
        * Z-확률변량에서 모분산을 표본분산으로 바꾸어 다음과 같은 확률변량을 만들 수 있다.
            * 이때의 자유도는 간단하게는 n-1과 m-1중 작은것을 사용하고, 더 정확하게는 등가자유도 df 값 사용
### t = $\frac{(\bar X_n - \bar Y_n) - (\mu_X - \mu_Y)}{(\frac{S^2_X}{n} + \frac{S^2_Y}{m})^\frac{1}{2}}$

이 구간추정은 정규분포에 바탕을 두고 있으므로, 모집단의 분포가 정규분포가 아닌 경우에는 표본의 크기가 커야 하는데, <br>
두 표본의 크기의 합이 약 40을 넘으면 정규성 가정에는 큰 무리가 없고, <br>
특별히 두 표본의 크기가 같을 때는 모집단의 분포에 많은 영향을 받지 않는다. <br>

[ 대응짝 표본 ]
예제 7.9 처럼 A브랜드 차량, B브랜드 차량과 같이 독립적인 경우가 아닌 동질적인 실험단위들을 짝지어 비교하는 경우, <br>
예를들어 한 사람의 다이어트 프로그램 시작 전 측정결과와 끝난 후 측정결과, 두 개의 관측값이 짝지어질 경우. 이때의 신뢰구간은 아래와 같다.<br>
### $\bar Y \pm t_{\frac{\alpha}{2}} (n-1) \frac{S_Y}{\sqrt n}$

### 7.6 모비율의 차에 대한 신뢰구간

=> 일반적으로 모비율의 차에 대한 정규근사에 있어서는, 일표본의 경우보다 작은 표본크기가 요구된다.

**예제) 유입경로별 기사박스 연속소비** <br>
전체 사용자 400명 중 기사박스(텍스트) 에서 유입된 사용자는 49%, 기사박스(이미지)에서 유입된 사용자는 44%, 핫토픽에서 유입된 사용자는 26%라고 할때 ,<br>
각 유입경로별 사용자 비율에 차이가 있다고 할 수 있는지 95% 신뢰구간을 구하라. <br>
(두 그룹씩 묶어서 비교)

### $\hat P_X$ - $\hat P_Y$ $\pm Z_\frac{\alpha}{2} \sqrt {(\frac{\hat P_X(1-\hat P_X)}{n} + \frac{\hat P_Y(1-\hat P_Y)}{m})}$

#### [1] 기사박스(텍스트) 와 기사박스(이미지) 비교
$\hat P_X$ = 0.49 이고,$\hat P_Y$ = 0.44, $Z_\frac{\alpha}{2}$ = 1.96 , n = 400 이라고 했을 때 <br>
위 공식에 대입하면, <br>

#### 0.49-0.44 $\pm$ 1.96 $\sqrt{\frac{0.49*0.51}{400}+\frac{0.44*0.56}{400}}$ = (-0.019,0.119) 
따라서, 95%신뢰수준이 -1.9% 에서 11.9% 로 신뢰구간이 0을 포함하므로, 기사박스(텍스트) 사용자 비율이(49%) 기사박스(이미지)(44%) 보다 크다고 할 수 없다.

#### [2] 기사박스(이미지) 와 핫토픽 비교
$\hat P_Y$ = 0.44 이고, $\hat P_Z$ = 0.26 , $Z_\frac{\alpha}{2}$ = 1.96 , n = 400 이라고 했을 때 <br>

#### 0.44-0.26 $\pm$ 1.96 $\sqrt{\frac{0.44*0.56}{400}+\frac{0.26*0.74}{400}}$ = (0.115,0.245)
따라서, 95%신뢰수준에서 11.5% 에서 24.5%로 기사박스(이미지)의 사용자 비율이(44%) 핫토픽(26%) 보다  더 크다.

```python
import numpy as np
def conf(p1, p2, n):
    min_conf = round(p1-p2 - 1.96*np.sqrt( (p1*(1-p1))/n + (p2*(1-p2))/n ),3)
    max_conf = round(p1-p2 + 1.96*np.sqrt( (p1*(1-p1))/n + (p2*(1-p2))/n ),3)
    
    print('interval:',min_conf, max_conf)
```

```python
# 기사박스(텍스트)= 0.49 , 기사박스(이미지)= 0.44 비교
conf(0.49,0.44,400)
```

```python
# 기사박스(이미지)= 0.44, 핫토픽 = 0.26 비교
conf(0.44,0.26,400)
```
