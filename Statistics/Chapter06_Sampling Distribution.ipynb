#### 관련 용어 정리
* 모집단(population) : 연구에서 실제 관심 있는 집단
    * ex) 전국의 모든 근로자, 전국의 모든 유권자
* 표본(sample) : 모집단의 일부분으로서 실제로 연구자가 자료를 수집하여 연구하는 부분
* 모수(parameter) : 모집단의 평균이나 분산, 비율이나 상관계수 등 여러가지 특성들
    * 모수는 모집단의 특성이므로 전체를 모두 관측할 수 없는 상황에서 정확하게 알 수 없으므로 통계적 추론에서 모수는 미지의 고정된 상수로 취급
* 통계량 (statistic) : 표본의 특성을 수치적으로 나타냄. 통계량은 특정한 표본이 주어지면 그 값을 완전히 결정할 수 있는 표본의 함수인데, 표본이 바뀌면 그 값도 바뀌므로 확률변수가된다.

## 6.1 통계량의 분포
### 6.1.1 표본분포
통계량의 표본분포란, 확률변량인 통계량의 확률분포를 말한다. <br>
즉, 모집단에서 표본의 크기가 n인 모든 표본이 뽑혔다고 가졍했을 때, 각 표본에서 계산된 통계량이 가지는 값들의 분포를 말한다. <br>

ex) 모집단에 속하는 모든 사람들의 평균 나이가 35세라고 할 때, 그 중에서 10명을 랜덤하게 뽑아 그들 나이의 평균을 계산하는 작업을 반복한다고 해 보자.<br>
어떤 표본의 평균은 37.3세, 33.6세 이런식으로 다르게 나타날 것이다. 이렇게 계산된 표본평균들을 다 모아 히스토그램을 그려보면<br>
표본평균은 모집단의 평균을 중심으로 분포하고 있음을 알 수 있다. <br>

* 표본평균의 표준편차는 모집단의 표준편차보다 작으며, 표본의 크기가 증가함에 따라 $\frac{1}{\sqrt n}$의 비율로 줄어들게 된다. 
* 즉, 평균이 $\mu$ 이고 표준편차가 $\sigma$인 모집단에서 크기 n의 임의 표본을 뽑아 표본평균을 구하면 $E(\bar X) =\mu$ 이고 $SD(\bar X) = \frac{\sigma}{\sqrt n}$이 된다는 것이다.

### 6.1.2 정규분포에서의 표본분포
정규분포로부터 얻은 랜덤표본의 함수로 표현되는 확률변량들의 분포에 대해 살펴보겠다. <br>

#### 모집단이 하나인 경우 :
모집단의 분포에 상관없이 표본이 랜덤표본이기만 하면 표본평균의 기댓값이 $\mu$이고 분산이 $\frac{\sigma ^2}{n}$ 임을 알 수 있다.

* **카이제곱분포**
### $\frac{(n-1)S^2}{\sigma ^2}$ = $\sum_{i=1}^n \frac{(X_i - \bar X_n)^2}{\sigma ^2}$ 
위의 확률변량은 자유도가 n-1인 카이제곱분포를 따름.

    * 카이제곱이라는 이름에는 표준정규분포를 제곱한다는 의미가 내포되어 있음.
    * 실제로 자유도 $v$ 인 카이제곱분포를 $X^2$을  $v$개 합한 것의 분포로 정의한다. 
    * 정확히 말하면 $X_1, X_2, ..., X_v$가 표준정규분포를 따를 때 $Q = X_1^2+X_2^2+X_3^2 ... +X_v^2$ 의 분포는 자유도가 $v$인 카이제곱 분포를 따른다고 정의한다.

[카이제곱분포 활용]
* 표본을 가지고 모분산을 추론하거나, 분포의 차이 등을 알아내고자 할 때

* **t 분포**
### $T = \frac{(\bar X_n - \mu)}{\frac{S}{\sqrt n}}$
위의 확률변량은 표본평균 $\bar X_n$을 표준화함에 있어서 미지의 표준편차 $\sigma$를 S로 대치한 결과이다.
    * 이러한 T는 자유도가 (n-1)인 t분포를 따른다.
        * 이유 : t분포의 자유도는 크기가 N인 표본에서 모분산이 추정되어야 한다는 제한을 받기 때문에 각 표본에서는 N-1개의 자유도를 가지게 된다.
    * t 확률변수의 분포형태는 0에 대하여 대칭이며 모수인 자유도에 의해 확률밀도함수의 꼬리 부분의 두터운 정도가 결정된다.
    * 표본의 크기 n이 크면 S가 $\sigma$의 가까운 추정값이 되므로, t분포는 표준정규분포와 매우 비슷한 분포가 되리라 짐작할 수 있다.
    * 표본으로부터 모집단의 모수들을 추정할 때, 모표준편차를 모르고 표본표준편차만 알고 있을 때 위의 식을 쓴다.

[t분포 활용]
* 정규분포의 평균을 측정할 때 주로 사용
* 모집단의 분산 (혹은 표준편차) 이 알려져 있지 않은 경우에 정규분포 대신 이용하는 확률분포.
* 두 집단간 평균 차이에 대한 가설검정의 경우 주로 사용.

#### 모집단이 두개인 경우 :

* **F분포**
### $F = \frac{S_X^2 / \sigma_X^2}{S_Y^2 / \sigma_Y^2}$
은 자유도가 (n-1, m-1)인 F분포를 따른다. 

[F분포 활용]
* 이러한 표본분포들은 모평균의 차이와 모분산의 비에 대한 통계적 추론에 활용된다.

## 6.2 중심극한정리
평균이 $\mu$ 이고 표준편차가 $sigma$인 모집단에서 랜덤표본을 뽑을 때, 표본크기 n이 충분히 크면 표본평균의 표본분포는 근사적으로 
### $\bar X_n$ ~ $N(\mu, \frac{\sigma ^2}{n})$

즉, 알 수 없는 모집단에서 표본이 충분히 크다면 이 표본평균의 분포는 정규분포에 근사한다는 것.
