# 실전 예측 분석 모델링

## **5장. 회귀 모델 성능 측정**
### 5.1 성능의 정량적 측정
* 결과가 수치형인 경우, 일반적으로 제곱근 평균 제곱 오차(root means squared error, RMSE) 를 사용한다. 
이 지표는 관측값에서 모델 예측값을 뺀 잔차에 대한 함수다.
* 평균 제곱 오차(mean squared error, MSE)는 잔차를 제곱해 합한 값.
    * 잔차값(의 평균)이 0으로부터 얼마나 떨어졌는지 또는 관측값과 모델의 예측값의 평균 거리가 얼마나 되는지를 파악하기 위함.

#### 결정계수 $R^2$
* $R^2$ 라고 쓰는 결정계수는 모델에 쓰이는 데이터의 정보 비율을 파악할 수 있다.
    * $R^2$가 0.75 라는 것은 이 모델은 결과 분포의 3/4을 설명할 수 있다는 뜻이다.
        * 이값을 구하는 가장 단순한 방법은 관측값과 예측한 값 간의 상관계수를 구한 후 제곱하는 것. <br>
        
※ 주의 : $R^2$ 는 상관값을 구하는 것이지, 정확도를 구하는 것이 아니라는 것을 이해해야 함.

* 완전히 동일한 RMSE를 갖는 다른 테스트 세트가 있지만, 그 테스트 세트 결과값의 분산이 더 작다면, 결과는 더 니쁘게 보일 수 있다. 
    * ex) 테스트 세트 결과의 분산이 4.2이고 RMSE가 1일 경우 $R^2$는 76% 정도이다. 동일한 RMSE를 갖는 다른 테스트세트의 분산이 3이라면 $R^2$는 67%일 것이다. <br>
        즉, 분산에 따라 RMSE가 같아도 $R^2$값은 달라질 수 있다.
* 모델의 핵심이 예측 정확도보다는 순위 분별력일 경우, 스피어만 순위 상관계수를 사용한다.
    * 스피어만 순위 상관계수 : 두 변수가 정규성을 따르지 않는 경우, 피어슨 상관계수 대신 사용. 값에 순위를 매겨 그 순위에 대해 상관계수 값을 구하기 때문에 비모수적 방법이라고 할 수 있다.

### 5.2 분산-편향성 트레이드 오프
* MSE의 기댓값 E 
    * E[MSE] = $\sigma^2$ + (모델편향성)^2 + 모델 분산
        * 여기서 첫항($\sigma^2$)은 보통 "기약 오차"라고 하는데, 이는 모델링에서 제거할 수 없다.
        * 두 번째 항은 모델 편향의 제곱값이다. 이는 모델의 함수 형식이 실제 예측 변수와 결과 간의 관계를 얼마나 제대로 표현했는지를 반영한다.
        * 마지막 항은 모델 분산이다.
* 편향 : 예측값들과 정답이 대체로 멀리 떨어져 있으면 결과의 편향(bias)이 높다고 말하고
    * 단순한 모델은 과적합이 생기지 않지만, 적합도가 너무 떨어지고 높은 편향성을 갖게 된다.
* 분산 : 예측값들이 자기들끼리 대체로 멀리 흩어져있으면 결과의 분산(variance)이 높다고 말한다.
    * 보통 보다 복잡한 모델일수록 높은 분산을 갖게되고, 이런 경우 과적합이 발생할 수 있다.
